```python
import yfinance as yf
import pandas as pd
import plotly.express as px
```

```python
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="1y")  # Último año de datos
tesla_data.reset_index(inplace=True)  # Para que la fecha sea una columna
tesla_data.head()  # Visualiza las primeras filas
```

```python
fig = px.line(tesla_data, x='Date', y='Close', title='Precio de cierre de Tesla')
fig.show()
```

```python
# Descargar datos históricos de GameStop
gamestop = yf.Ticker("GME")
gamestop_data = gamestop.history(period="1y")  # Último año de datos

# Asegúrate de que el índice de la fecha esté como columna para facilitar el análisis
gamestop_data.reset_index(inplace=True)

# Mostrar las primeras filas del DataFrame para revisar los datos
gamestop_data.head()
```

```python
import plotly.express as px

fig = px.line(gamestop_data, x='Date', y='Close', title='Precio de cierre de GameStop')
fig.show()
```

```python
gamestop = yf.Ticker("GME")
gamestop_data = gamestop.history(period="1y")
gamestop_data.reset_index(inplace=True)

# Agregar etiquetas para diferenciar las acciones
tesla_data['Stock'] = 'Tesla'
gamestop_data['Stock'] = 'GameStop'

# Combinar ambos DataFrames
combined_data = pd.concat([tesla_data, gamestop_data])

# Graficar ambas acciones
fig = px.line(combined_data, x='Date', y='Close', color='Stock', title='Comparación de Tesla y GameStop')
fig.show()
```
